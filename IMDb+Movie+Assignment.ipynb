-  ###  Subtask 3.1 Combine the Dataframe by Genres

There are 3 columns in the dataframe - `genre_1`, `genre_2`, and `genre_3`. As a part of this subtask, you need to aggregate a few values over these 3 columns. 
1. First create a new dataframe `df_by_genre` that contains `genre_1`, `genre_2`, and `genre_3` and all the columns related to **CVotes/Votes** from the `movies` data frame. There are 47 columns to be extracted in total.
2. Now, Add a column called `cnt` to the dataframe `df_by_genre` and initialize it to one. You will realise the use of this column by the end of this subtask.
3. First group the dataframe `df_by_genre` by `genre_1` and find the sum of all the numeric columns such as `cnt`, columns related to CVotes and Votes columns and store it in a dataframe `df_by_g1`.
4. Perform the same operation for `genre_2` and `genre_3` and store it dataframes `df_by_g2` and `df_by_g3` respectively. 
5. Now that you have 3 dataframes performed by grouping over `genre_1`, `genre_2`, and `genre_3` separately, it's time to combine them. For this, add the three dataframes and store it in a new dataframe `df_add`, so that the corresponding values of Votes/CVotes get added for each genre.There is a function called `add()` in pandas which lets you do this. You can refer to this link to see how this function works. https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.DataFrame.add.html
6. The column `cnt` on aggregation has basically kept the track of the number of occurences of each genre.Subset the genres that have atleast 10 movies into a new dataframe `genre_top10` based on the `cnt` column value.
7. Now, take the mean of all the numeric columns by dividing them with the column value `cnt` and store it back to the same dataframe. We will be using this dataframe for further analysis in this task unless it is explicitly mentioned to use the dataframe `movies`.
8. Since the number of votes can't be a fraction, type cast all the CVotes related columns to integers. Also, round off all the Votes related columns upto two digits after the decimal point.


```python
# Create the dataframe df_by_genre
w1 = ['Votes','genre']
col_list = [x for x in list (movies.columns.unique()) if (w1[0] in x) or (w1[1] in x)]
df_by_genre = movies[col_list]
print('Shape of df_by_genre:', df_by_genre.shape)
df_by_genre.head(2)
```

```python
# Create a column cnt and initialize it to 1
df_by_genre['cnt'] = 1
df_by_genre.tail(2)
```

```python
# Group the movies by individual genres
df_by_g1 = df_by_genre.groupby('genre_1').sum()
df_by_g2 = df_by_genre.groupby('genre_2').sum()
df_by_g3 = df_by_genre.groupby('genre_3').sum()
print (df_by_g1.shape, df_by_g2.shape, df_by_g3.shape)
df_by_g1.head(4)
#df_by_g2.head(4)
#df_by_g3.head(4)
```

```python
# Add the grouped data frames and store it in a new data frame
df_add=df_by_g1.add(df_by_g2,fill_value=0).add(df_by_g3,fill_value=0)
print ('CVotes10 Manual Addition for genre Drama', df_by_g1.loc['Drama'].CVotes10+df_by_g2.loc['Drama'].CVotes10+df_by_g3.loc['Drama'].CVotes10)
print ('CVotes10 Add Function for genre Drama   ', df_add.loc['Drama'].CVotes10)
```

```python
# Extract genres with atleast 10 occurences
genre_top10=df_add[df_add['cnt']>=10]
genre_top10
```

```python
# Take the mean for every column by dividing with cnt 
for i in range(genre_top10.shape[1]-1):
    genre_top10.iloc[:,i] = genre_top10.iloc[:,i] / genre_top10['cnt']
genre_top10
```

```python
# Rounding off the columns of Votes to two decimals
start_index=genre_top10.columns.get_loc("VotesM")
end_index=genre_top10.columns.get_loc("VotesnUS")+1
genre_top10.iloc[:,start_index:end_index]=genre_top10.iloc[:,start_index:end_index].round(decimals=2)
```

```python
# Converting CVotes to int type
start_index=genre_top10.columns.get_loc("CVotes10")
end_index=genre_top10.columns.get_loc("CVotesnUS")+1
genre_top10.iloc[:,start_index:end_index]=genre_top10.iloc[:,start_index:end_index].astype('int64').convert_dtypes()
genre_top10
```

```python
movies.Country.value_counts()
```

```python
# Creating IFUS column
movies['IFUS'] = movies.Country.apply(lambda x: 'USA' if x == 'USA' else 'non-USA')
movies.IFUS.value_counts()
```

```python
# Box plot - 1: CVotesUS(y) vs IFUS(x)
plt.figure(figsize=[18,7])
plt.style.use('ggplot')

plt.subplot(1,2,1)
sns.boxplot(data=movies,x='IFUS',y='CVotesUS',fliersize=7)
plt.title('Box Plot of US Vote Count for USA & non-USA movies', fontdict={'fontsize':15, 'color':'black'})
plt.xlabel('IFUS', fontdict = {'fontsize': 13, 'fontweight': 6, 'color' : 'black'})
plt.ylabel('CVotes US', fontdict = {'fontsize': 13, 'fontweight': 6, 'color' : 'black'})
ticks = np.arange(0,250000,25000)
labels = ['{}K'.format(int(x/1000)) for x in ticks]
plt.yticks(ticks,labels)

plt.subplot(1,2,2)
sns.boxplot(data=movies,x='IFUS',y='CVotesnUS',fliersize=7)
plt.title('Box Plot of non-US Vote Count for USA & non-USA movies', fontdict={'fontsize':15, 'color':'black'})
plt.xlabel('IFUS', fontdict = {'fontsize': 13, 'fontweight': 6, 'color' : 'black'})
plt.ylabel('CVotes non-US', fontdict = {'fontsize': 13, 'fontweight': 6, 'color' : 'black'})
ticks = np.arange(0,800000,50000)
labels = ['{}K'.format(int(x/1000)) for x in ticks]
plt.yticks(ticks,labels)

plt.show()
```

**`Inferences:`** Write your two inferences/observations below:
- Inference 1: Non USA movies tend to have more spread in terms of number of votes (higher IQR) whether the votes are coming from USA or non USA audience. Also note that there is no outliers based on the standard 1.5 * IQR when it comes to non-USA movies. It needs a closer look to determine if this is because the sample size of non-USA movies is much lower than USA movies. 
- Inference 2: Non USA vote count appears to outnumber the US vote count. Despite this, the US still movies have a slighty higher median vote count. Does this mean US movies are more popular than non-US movies ? Well need to look at the score of the votes or rating values.

```python
# Box plot - 2: VotesUS(y) vs IFUS(x)

plt.figure(figsize=[18,7])
plt.style.use('ggplot')

plt.subplot(1,2,1)
sns.boxplot(data=movies,x='IFUS',y='VotesUS',fliersize=7)
plt.title('Box Plot of US Ratings for USA & non-USA movies', fontdict={'fontsize':15, 'color':'black'})
plt.xlabel('IFUS', fontdict = {'fontsize': 13, 'fontweight': 6, 'color' : 'black'})
plt.ylabel('Votes US', fontdict = {'fontsize': 13, 'fontweight': 6, 'color' : 'black'})
ticks = np.arange(7.2,9.2,0.2)
plt.yticks(ticks)

plt.subplot(1,2,2)
sns.boxplot(data=movies,x='IFUS',y='VotesnUS',fliersize=7)
plt.title('Box Plot of non-US Ratings for USA & non-USA movies', fontdict={'fontsize':15, 'color':'black'})
plt.xlabel('IFUS', fontdict = {'fontsize': 13, 'fontweight': 6, 'color' : 'black'})
plt.ylabel('Votes non-US', fontdict = {'fontsize': 13, 'fontweight': 6, 'color' : 'black'})
ticks = np.arange(7.2,9.2,0.2)
plt.yticks(ticks)

plt.show()
```

**`Inferences:`** Write your two inferences/observations below:
- Inference 1: Median of the ratings for USA movies is higher than non-USA movies. This trend remains the same, irrespective of whether the rating is from US or non-US audience.
- Inference 2: Ratings from US audience is a notch higher than ratings from non-US audience. This trend is the same irrespective of whether the movie is from US or non-US. In addition, the spread of ratings for non-USA movies rated by non-USA audience is little higher. Perhaps a more segmented analysis within non-USA movie category of ISUS variable would throw more light into this aspect.
